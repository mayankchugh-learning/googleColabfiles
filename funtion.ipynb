```python
def test7(a):
  n = []
  for i in a :
    if type(i) == list:
      for j in i:
        if type(j) == int or type(j) == float:
          n.append(j)
    else:
          if type(i) == int or type(i) == float:
            n.append(i)
  return n
```

```python
def test14(**kwargs):
  for i in kwargs.keys():
    if type(kwargs[i]) == list:
      return i, kwargs[i]
```

```python
test14(a=34,b=23,c=[1,2,3,4],d=("sudh","skills"))
```
