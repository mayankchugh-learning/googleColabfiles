# Comparision Operation:



```
Operation           Meaning
---------------------------------------
<                   less than
<=             less than or equal to
<                   greter than
<=             greater than or equal to
```

````python
maximum_speed_of_bike = 150
maximum_speed_of_car = 200

print(f"bike is faster than car: {maximum_speed_of_bike > maximum_speed_of_car}")
print(f"bike is faster than car: {maximum_speed_of_bike >= maximum_speed_of_car}")
print(f"bike is faster than car: {maximum_speed_of_bike < maximum_speed_of_car}")
print(f"bike is faster than car: {maximum_speed_of_bike <= maximum_speed_of_car}")
````
